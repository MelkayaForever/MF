```python
import pandas as pd
```

```python
reestr = pd.read_excel('/content/drive/MyDrive/ДЗ.xlsx', sheet_name='Реестр')
sprav = pd.read_excel('/content/drive/MyDrive/ДЗ.xlsx', sheet_name='Справочник')
reestr
```

```python
reestr.info()
```

```python
reestr['DE_year'] = reestr['DE'].astype(str).str.split("-").str[0]
reestr
```

```python
forms =  reestr.groupby('Форма').groups
forms
```

```python
forms1 = reestr.groupby('Форма', group_keys=False).apply(lambda x: x)
forms1
```

Для заполнения пустых значений в 'Форма' не хватает данных, но предположим, что существующее и по сей день (DE_year = 2999) будет НАО

```python
reestr.loc[reestr['DE_year']=='2999', 'Форма'] = 'НАО'
```

```python
reestr['Форма'].fillna('ЗАО', inplace=True)
```

ID_City, где 'Наименование' = Зимний дворец, заполним по образцу

```python
reestr['ID_City'] = reestr['ID_City'].fillna(method ='bfill')
```

```python
reestr
```

```python
sprav
```

```python
sprav_copy = sprav
```

```python
sprav_copy.info()
```

Для Москвы добавим код 77, для Санкт-Петербурга 78. Также, заменим значения Петроград и Ленинград на Санкт-Петербург

```python
sprav_copy.loc[sprav_copy['ID']==2, 'Код'] = 78
```

```python
sprav_copy.loc[sprav_copy['ID']==2, 'Название'] = 'Санкт-Петербург'
```

```python
sprav_copy.loc[sprav_copy['ID']==1, 'Код'] = 77
```

```python
sprav_copy
```

Оставим в таблице только нужные данные, сгруппируем по ID, убрав его в индекс для дальнейшего объединения

```python
sprav_copy = sprav_copy[['ID', 'Код', 'Название']].groupby('ID').first()
sprav_copy
```

Объединим таблички, затем оставим только нужные колонки

```python
df= reestr.merge(sprav_copy, how='left', left_on='ID_City', right_on='ID')
```

```python
df
```

```python
df_cb = df[['Наименование', 'Форма', 'ID_City', 'Название']]
df_cb
```
